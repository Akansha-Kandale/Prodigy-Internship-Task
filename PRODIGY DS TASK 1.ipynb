```python
#Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Choosing a country or region to visualize
country_to_visualize = 'World'
```

```python
#Filtering data for the selected country or region
selected_country_data_total = total[total['Country Name'] == country_to_visualize]
selected_country_data_male = male_population[male_population['Country Name'] == country_to_visualize]
selected_country_data_female = female_population[female_population['Country Name'] == country_to_visualize]
```

```python
#Plotting a bar chart for the distribution of male and female populations
years = selected_country_data_total.columns[4:]
plt.figure(figsize=(12, 6))
plt.bar(years, selected_country_data_male.iloc[0, 4:], label='Male', color='#0000FF', alpha=0.7)
plt.bar(years, selected_country_data_female.iloc[0, 4:], label='Female', bottom=selected_country_data_male.iloc[0, 4:], color='#FF6EC7', alpha=0.7)

plt.title(f'Population Distribution in {country_to_visualize} by Gender (1960-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Filtering data for the selected country or region
selected_country_data_total = total[total['Country Name'] == country_to_visualize]
selected_country_data_male = male_population[male_population['Country Name'] == country_to_visualize]
selected_country_data_female = female_population[female_population['Country Name'] == country_to_visualize]

years = selected_country_data_total.columns[4:]
male_population = selected_country_data_male.iloc[0, 4:]
female_population = selected_country_data_female.iloc[0, 4:]
```

```python
# Assuming we have yearly data, we can calculate the number of years
num_years = len(years)

# Setting the bar width based on the number of years and our preference
bar_width = 0.35  # Adjust this value as needed

# Create a numeric representation of years for better x-axis labeling
years_numeric = [int(year) for year in years]
```

```python
# Plotting a bar chart for the distribution of male and female populations
plt.figure(figsize=(12, 6))
plt.bar(years_numeric, male_population, width=bar_width, label='Male', color='#0000FF', alpha=0.7)
plt.bar([year + bar_width for year in years_numeric], female_population, width=bar_width, label='Female', color='#FF6EC7', alpha=0.7)

plt.title(f'Population Distribution in {country_to_visualize} by Gender (1960-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks([year + bar_width/2 for year in years_numeric], years_numeric, rotation=90)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
